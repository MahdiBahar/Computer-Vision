# Image Enhancement

### import part

```python
import cv2 as cv

import matplotlib.pyplot as plt
import numpy as np
```

```python
def image_plot (img , width =6 , hight = 8 , c_map='gray' , title = 'image'):
    plt.figure(figsize=(hight,width),dpi=100)
    plt.axis('off')
    plt.imshow(img, cmap=c_map)
    plt.title(title)
    plt.show()
```

```python
org_img = cv.imread(r'images/original.jpg',cv.IMREAD_UNCHANGED)
```

```python
image_plot(org_img,width = 8, hight = 12 , title = 'orginal image')
```

```python
def display_hist (img , channel = 0, histsize = 256, range= (0,256), title = 'histogram'):

    hist_img = cv.calcHist([img],[channel],None,[histsize],range)
    plt.figure(figsize=(6,4),dpi = 100)
    plt.plot(hist_img)
    plt.title(title)
    plt.show()
```

```python
display_hist(org_img)
```

```python
def display_hist_bins(img, bins = 256, range=(0,256), title = 'histogram'):
    plt.figure(figsize= (6,4) , dpi=100)
    plt.hist(img.flat , bins , range)
    plt.title(title)
    plt.show()
```

```python
display_hist_bins(org_img)
```

# Contrast Enhancement

## Contrast Stretching

```python
min_new = 0
max_new = 255

min_old = np.min(org_img)
max_old = np.max(org_img)

cs_img = np.uint8(((org_img-min_old)/(max_old-min_old))*(max_new-min_new) + min_new)
```

```python
image_plot(cs_img,title = 'image after contrast stretching')
```

```python
display_hist_bins(cs_img, title = 'histogram after CS')
```

#### there is not much improvment after using contrast stretching

## Histogram Equalization

```python
histequal = cv.equalizeHist(org_img)
```

```python
image_plot(histequal,title= 'image after applying histogram equalization')
```

## Gamma

```python
gamma = 0.5
gamma_img = np.uint8(((org_img/255)**gamma)*255)
```

```python
image_plot(gamma_img,title = 'image after applying gamma')
```

## Combine Gamma with historam equalization

```python
histequal_combine = cv.equalizeHist(gamma_img)
```

```python
image_plot(histequal_combine, title = 'image after applying gamma and histogram equalization')
```

### The best result is created by histogram equalization
